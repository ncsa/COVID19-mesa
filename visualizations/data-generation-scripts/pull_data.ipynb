```python
async def get_step(websocket, step):
    await websocket.send(json.dumps({"type": "get_step", "step":str(step)}))
    print(step)
    return await websocket.recv()
```

```python
async def reset(websocket):
    await websocket.send('{"type":"reset"}')
    return await websocket.recv()
```
